```python
# calculate the factorial of number
n=int(input("enter the number"))
fact=1
for i in range(1,n+1):
    fact*=i
print(f"factorial of this number is {fact}")
```

```python
# check the prime number or not
n=int(input("enter the number"))
if n<2:
    print('Not prime')
else:
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            print("Not Prime")
            break
    else:
        print("prime")
    
```
