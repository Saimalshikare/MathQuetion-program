```python
# to check number is perfect number or not
n=28
divisor=[]
for i in range(1,n):
    if n%i==0:
        divisor.append(i)
print(f"divisor is {divisor}")
if sum(divisor)==n:
    print(f"{n} is a perfect number")
else:
    print(f"{n} is a not perect number")
    
```

```python
# to Check this number is strong number or not
import math
n=int(input("enter the number"))
total=0
for digit in str(n):
    total+=math.factorial(int(digit))
if total==n:
    print(f"{n} is Stong Number")
else:
    print(f"{n} is not a Strong Number")
```

```python
# Find all the prime between 1 to n
n = int(input("Enter the number: "))

for num in range(2, n+1):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```
